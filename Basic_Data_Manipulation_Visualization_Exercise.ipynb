### 49. Write a Python script to display the
- a. Current date and time
- b. Current year
- c. Month of year
- d. Week number of the year
- e. Weekday of the week
- f. Day of year
- g. Day of the month
- h. Day of week

```python
from datetime import datetime

# Get the current date and time
now = datetime.now()

# Display the required information
print("a. Current date and time:", now)
print("b. Current year:", now.year)
print("c. Month of year:", now.strftime('%B'))  # Full month name
print("d. Week number of the year:", now.strftime('%U'))  # Week number
print("e. Weekday of the week:", now.strftime('%A'))  # Full weekday name
print("f. Day of year:", now.timetuple().tm_yday)  # Day number in the year
print("g. Day of the month:", now.day)
print("h. Day of week:", now.weekday())  # 0 = Monday, 6 = Sunday
```

### 50. Write a Python program to convert a string to datetime.

### 54. Write a Python program to get days between two dates. 

```python
from datetime import datetime

# Define two dates as strings
date1 = "2024-11-28"
date2 = "2024-12-05"

# Convert the strings to datetime objects
date1_obj = datetime.strptime(date1, "%Y-%m-%d")
date2_obj = datetime.strptime(date2, "%Y-%m-%d")

# Calculate the difference in days
days_difference = (date2_obj - date1_obj).days

# Display the result
print(f"The number of days between {date1} and {date2} is {days_difference} days.")
```

### 80. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries
Hint: First we need to find top-3 countries based on revenue, then create scater plot between Quantity and Unitprice for each country separately


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
online_rt = pd.read_csv(r"C:\Users\Lenovo\Downloads\Python\Online_Retail.csv", encoding='ISO-8859-1')

# Calculate total revenue for each country
online_rt['Revenue'] = online_rt['Quantity'] * online_rt['UnitPrice']
country_revenue = online_rt.groupby('Country')['Revenue'].sum().sort_values(ascending=False)

# Get the top 3 countries by revenue
top_3_countries = country_revenue.head(3).index

# Filter data to include only rows from the top 3 countries
top_3_data = online_rt[online_rt['Country'].isin(top_3_countries)]

# Create scatter plots for each top 3 country
plt.figure(figsize=(18, 6))
for i, country in enumerate(top_3_countries, 1):
    plt.subplot(1, 3, i)  # Create subplots in a 1x3 grid
    country_data = top_3_data[top_3_data['Country'] == country]
    plt.scatter(country_data['Quantity'], country_data['UnitPrice'], alpha=0.5, label=country)
    plt.title(f'Scatter Plot of Quantity vs UnitPrice for {country}')
    plt.xlabel('Quantity')
    plt.ylabel('Unit Price')
    plt.grid(True)

plt.tight_layout()  # Adjust layout to fit subplots nicely
plt.show()
```

### 88. Calculate total sales data (quantity) for 2019 for each product and show it using a Pie chart. Understand percentage contribution from each product

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"C:\Users\Lenovo\Downloads\Python\FMCG_Company_Data_2019.csv"
company_data = pd.read_csv(file_path)

# Filter data for the year 2019
company_data['Month'] = pd.to_datetime(company_data['Month'], format='%b-%y')  # Convert 'Month' column to datetime
company_data_2019 = company_data[company_data['Month'].dt.year == 2019]

# Calculate total sales data (quantity) for each product
product_sales_2019 = company_data_2019[['FaceCream', 'FaceWash', 'ToothPaste', 'Soap', 'Shampo', 'Moisturizer']].sum()

# Plot a pie chart
plt.figure(figsize=(10, 8))
plt.pie(product_sales_2019, labels=product_sales_2019.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Add a title
plt.title('Percentage Contribution of Each Product to Total Sales in 2019')

# Display the pie chart
plt.show()
```
